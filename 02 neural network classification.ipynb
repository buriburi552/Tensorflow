```python
X_train = X[:800]
X_test = X[800:]

y_train = y[:800]
y_test = y[800:]
```

```python
tf.random.set_seed=42

final_model= tf.keras.Sequential()

final_model.add(tf.keras.layers.Dense(units=10, activation='relu'))
final_model.add(tf.keras.layers.Dense(units=10, activation='relu'))
final_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

final_model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])
```

```python
history_final = final_model.fit(X_train,y_train,epochs=70)
```

```python
final_model.evaluate(x=X_test,y=y_test)
```

```python
pred = final_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_test
```

```python
print(classification_report(np.array(y_test),np.array(pred)))

print(confusion_matrix(y_test,pred))
```

```python
tf.random.set_seed=42

model5 = tf.keras.Sequential()

model5.add(tf.keras.layers.Dense(units=5, activation='relu'))
model5.add(tf.keras.layers.Dense(units=5, activation='relu'))
model5.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

model5.compile(loss='binary_crossentropy',
               optimizer='Adam',
               metrics=['accuracy'])

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

history5 = model5.fit(X_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history5.history).plot(figsize=(10,7), xlabel='epochs');
```

```python
lrs = 1e-4 * (10 ** (tf.range(100)/20))
loss = pd.DataFrame(history5.history)['loss']
```
